```python
# Import required modules
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
from PyPDF2 import PdfReader
import gradio as gr
import fitz
from transformers import pipeline

# Initialize Stable Diffusion Pipeline
def load_pipeline():
    model_id = "CompVis/stable-diffusion-v1-4"
    pipe = StableDiffusionPipeline.from_pretrained(model_id)

    # Check for GPU and set device
    device = "cuda" if torch.cuda.is_available() else "cpu"
    pipe.to(device)

    # Use DPMSolverMultistepScheduler for better image generation
    pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
    return pipe

pipe = load_pipeline()

# Text-to-Image Generation Function
def generate_image(prompt, height=512, width=512, steps=50):
    try:
        # Generate image
        result = pipe(prompt, height=height, width=width, num_inference_steps=steps)
        image = result.images[0]
        return image
    except Exception as e:
        return f"Error: {e}"

# PDF Summarization Function
def summarize_pdf(pdf_filepath):
    try:
        # First attempt using PyPDF2
        text = ""
        reader = PdfReader(pdf_filepath)
        for page in reader.pages:
            text += page.extract_text()

        if not text.strip():  # Check if PyPDF2 failed to extract content
            raise ValueError("PyPDF2 failed to extract content. Switching to PyMuPDF.")

    except Exception as e:
        # Fallback to PyMuPDF
        try:
            text = ""
            pdf_document = fitz.open(pdf_filepath)
            for page in pdf_document:
                text += page.get_text()
            pdf_document.close()
        except Exception as fallback_error:
            return f"Error: Could not extract text from the PDF. Details: {fallback_error}"

    # Summarize the extracted text
    try:
        summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
        summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
        return summary[0]['summary_text']
    except Exception as summary_error:
        return f"Error during summarization: {summary_error}"

# Gradio Interface
def main_interface():
    with gr.Blocks() as app:
        gr.Markdown("# AI Model for Text-to-Image and PDF Summarization")

        # Text-to-Image Section
        with gr.Row():
            with gr.Column():
                prompt_input = gr.Textbox(label="Enter Text Prompt", placeholder="Describe the image you want...")
                height_input = gr.Slider(256, 768, value=512, step=16, label="Height")
                width_input = gr.Slider(256, 768, value=512, step=16, label="Width")
                #steps_input = gr.Slider(10, 100, value=50, step=5, label="Number of Steps")
                generate_button = gr.Button("Generate Image")
            with gr.Column():
                image_output = gr.Image(label="Generated Image")

        # Bind Text-to-Image Functionality
        generate_button.click(
            fn=generate_image,
            inputs=[prompt_input, height_input, width_input],
            outputs=image_output
        )

        gr.Markdown("## PDF Summarization")
        # PDF Summarization Section
        with gr.Row():
            with gr.Column():
                pdf_input = gr.File(label="Upload PDF", type="filepath")
                summarize_button = gr.Button("Summarize PDF")
            with gr.Column():
                summary_output = gr.Textbox(label="PDF Summary", lines=10)

        # Bind PDF Summarization Functionality
        summarize_button.click(
            fn=summarize_pdf,
            inputs=[pdf_input],
            outputs=summary_output
        )

    return app

# Launch the App
app = main_interface()
app.launch()
```

```python
# Imports
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
from transformers import pipeline
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
import fitz  # PyMuPDF
import gradio as gr

# Load Stable Diffusion
def load_pipeline():
    model_id = "CompVis/stable-diffusion-v1-4"
    pipe = StableDiffusionPipeline.from_pretrained(model_id)
    pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
    device = "cuda" if torch.cuda.is_available() else "cpu"
    pipe.to(device)
    return pipe

pipe = load_pipeline()

# Text-to-Image Function
def generate_image(prompt, height=512, width=512):
    try:
        result = pipe(prompt, height=height, width=width, num_inference_steps=50)
        return result.images[0]
    except Exception as e:
        return f"Error: {e}"

# -------- RAG-based PDF Q&A --------
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Extract text chunks from PDF
def extract_chunks(pdf_path, chunk_size=500):
    text = ""
    doc = fitz.open(pdf_path)
    for page in doc:
        text += page.get_text()
    doc.close()

    # Split into chunks
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

# Create FAISS index from chunks
def create_faiss_index(chunks):
    embeddings = embedding_model.encode(chunks)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(np.array(embeddings))
    return index, chunks, embeddings

# Answer a question based on PDF
def answer_question_over_pdf(pdf_path, user_question, top_k=3):
    try:
        chunks = extract_chunks(pdf_path)
        index, stored_chunks, _ = create_faiss_index(chunks)

        # Embed the user question
        question_embedding = embedding_model.encode([user_question])
        D, I = index.search(np.array(question_embedding), top_k)

        retrieved_chunks = [stored_chunks[i] for i in I[0]]
        context = " ".join(retrieved_chunks)

        # Generate answer from retrieved context
        answer = summarizer(context + " Question: " + user_question,
                            max_length=130, min_length=30, do_sample=False)
        return answer[0]['summary_text']
    except Exception as e:
        return f"Error: {e}"

# Gradio App
def main_interface():
    with gr.Blocks() as app:
        gr.Markdown("# AI Tool: Text-to-Image + Ask Questions from PDFs")

        # Text-to-Image
        gr.Markdown("## 🎨 Text-to-Image Generator")
        with gr.Row():
            with gr.Column():
                prompt_input = gr.Textbox(label="Enter Prompt", placeholder="A futuristic city at night")
                height_input = gr.Slider(256, 768, value=512, step=16, label="Image Height")
                width_input = gr.Slider(256, 768, value=512, step=16, label="Image Width")
                generate_button = gr.Button("Generate Image")
            with gr.Column():
                image_output = gr.Image(label="Generated Image")

        generate_button.click(
            fn=generate_image,
            inputs=[prompt_input, height_input, width_input],
            outputs=image_output
        )

        # PDF Q&A
        gr.Markdown("## 📄 Ask Questions from PDF")
        with gr.Row():
            with gr.Column():
                pdf_input = gr.File(label="Upload PDF", type="filepath")
                question_input = gr.Textbox(label="Your Question", placeholder="E.g., What is the objective?")
                ask_button = gr.Button("Get Answer")
            with gr.Column():
                answer_output = gr.Textbox(label="Answer", lines=10)

        ask_button.click(
            fn=answer_question_over_pdf,
            inputs=[pdf_input, question_input],
            outputs=answer_output
        )

    return app

# Launch App
app = main_interface()
app.launch()
```

```python
# Step 1: Install necessary libraries
!pip install -q transformers sentence-transformers faiss-cpu PyMuPDF

# Step 2: Import necessary libraries
import fitz  # PyMuPDF for PDF text extraction
import faiss
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from google.colab import files

# Step 3: Upload PDF file
uploaded = files.upload()

# Extract text from PDF
def extract_text_from_pdf(path):
    doc = fitz.open(path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Step 4: Split text into chunks (for efficient retrieval)
def split_text(text, chunk_size=300):
    words = text.split()
    return [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]

# Step 5: Load Sentence-BERT and create embeddings for chunks
embedder = SentenceTransformer("all-MiniLM-L6-v2")

# Step 6: Create FAISS index for text chunks
def create_faiss_index(chunks):
    corpus_embeddings = embedder.encode(chunks, convert_to_tensor=True)
    corpus_embeddings_np = corpus_embeddings.cpu().detach().numpy()
    dimension = corpus_embeddings_np.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(corpus_embeddings_np)
    return index, chunks

# Step 7: Function to answer questions from PDF using transformer model
def answer_question(question, index, chunks):
    q_embedding = embedder.encode(question, convert_to_tensor=True)
    q_embedding_np = q_embedding.cpu().detach().numpy()

    # Retrieve top 3 relevant chunks
    top_k = 3
    _, indices = index.search(q_embedding_np.reshape(1, -1), top_k)
    retrieved_context = "\n".join([chunks[i] for i in indices[0]])

    # Use transformer model for QA
    qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
    result = qa_pipeline({'context': retrieved_context, 'question': question})

    return result['answer']

# Step 8: Main function to upload PDF, process it, and answer questions
def process_pdf_and_ask_questions():
    # Get uploaded PDF
    pdf_path = list(uploaded.keys())[0]
    full_text = extract_text_from_pdf(pdf_path)

    # Split text into chunks
    chunks = split_text(full_text)

    # Create FAISS index for similarity search
    index, chunks = create_faiss_index(chunks)

    # Loop for multiple questions
    while True:
        # Ask user for question
        question = input("\nAsk a question (or type 'exit' to quit): ")
        if question.lower() == 'exit':
            print("Exiting...")
            break

        # Get the answer for the question
        answer = answer_question(question, index, chunks)

        # Display the answer
        print(f"Answer: {answer}")

# Run the function to upload PDF and interact with the user
process_pdf_and_ask_questions()
```

```python
# Required Libraries
from transformers import pipeline
from PyPDF2 import PdfReader
import fitz  # PyMuPDF
import gradio as gr
import os

# Use transformer QA model to answer question from retrieved text
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# PDF Text Extraction Function using PyPDF2 and PyMuPDF (fallback)
def extract_pdf_text(pdf_filepath):
    try:
        # First attempt using PyPDF2
        text = ""
        reader = PdfReader(pdf_filepath)
        for page in reader.pages:
            text += page.extract_text()

        if not text.strip():  # Check if PyPDF2 failed to extract content
            raise ValueError("PyPDF2 failed to extract content. Switching to PyMuPDF.")

    except Exception as e:
        # Fallback to PyMuPDF
        try:
            text = ""
            pdf_document = fitz.open(pdf_filepath)
            for page in pdf_document:
                text += page.get_text()
            pdf_document.close()
        except Exception as fallback_error:
            return f"Error: Could not extract text from the PDF. Details: {fallback_error}"

    return text

# Function to chunk large context into smaller pieces
def chunk_text(text, chunk_size=1000):
    # Split the text into chunks of specified size (e.g., 1000 characters)
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

# Function to answer questions from the context
def answer_question(context, question):
    # Ensure context is not too short
    if len(context) < 10:
        return "Error: Context too short for answering."

    # Ensure truncation if the context is too long (if it exceeds max length)
    context = context[:1000]  # Limit context to 1000 tokens (adjust as needed)

    try:
        result = qa_pipeline({'context': context, 'question': question})
        return result['answer']
    except Exception as e:
        return f"Error during question answering: {e}"

# Main function for processing PDF and answering questions
def process_pdf_and_answer_questions(pdf_file, question):
    # Step 1: Extract text from the PDF
    extracted_text = extract_pdf_text(pdf_file.name)

    if "Error" in extracted_text:
        return extracted_text  # If there's an error in extraction, return the error message

    # Step 2: Chunk the extracted text (if necessary)
    chunks = chunk_text(extracted_text)

    # Step 3: Process each chunk to get the answer from each piece
    answers = []
    for chunk in chunks:
        answer = answer_question(chunk, question)
        if answer != "Error: Context too short for answering.":
            answers.append(answer)

    # Combine answers from all chunks
    return " ".join(answers) if answers else "No answer found."

# Gradio Interface
def gradio_interface():
    with gr.Blocks() as app:
        gr.Markdown("# AI Model for PDF Question-Answering")

        # PDF Upload Section
        with gr.Row():
            with gr.Column():
                pdf_input = gr.File(label="Upload PDF", type="filepath")  # Fixing 'type' to 'filepath'
                question_input = gr.Textbox(label="Enter Question", placeholder="Ask a question about the PDF content")
                submit_button = gr.Button("Get Answer")
            with gr.Column():
                answer_output = gr.Textbox(label="Answer", lines=5)

        # Bind PDF Question-Answering
        submit_button.click(
            fn=process_pdf_and_answer_questions,
            inputs=[pdf_input, question_input],
            outputs=answer_output
        )

    return app

# Launch the App
app = gradio_interface()
app.launch()
```
